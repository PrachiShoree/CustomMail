```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/Tailor AI/Student Data Updated.csv')

# Count occurrences of each niche profession
profession_count = df['Niche Profession'].value_counts()

# Prepare data for the step plot
labels = profession_count.index
counts = profession_count.values

# Create a step plot
plt.figure(figsize=(10, 6))
plt.step(labels, counts, where='mid', color='orange', linewidth=2)

# Adding markers for each data point
plt.scatter(labels, counts, color='blue', zorder=5)

# Set title and labels
plt.title('Niche Profession Distribution (Step Plot)', fontsize=16)
plt.xlabel('Niche Profession', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
