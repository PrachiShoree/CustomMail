```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
# Replace 'your_file.csv' with the actual path or name of the uploaded file
df = pd.read_csv('/content/drive/MyDrive/Tailor AI/intern_data Updated.csv')

# Preview the data to understand its structure
print(df.head())  # This will print the first 5 rows of the dataset
```

```python
#  Niche Profession Distribution
plt.figure(figsize=(10, 6))  # Set the figure size

# Create a count plot for niche profession distribution
sns.countplot(y='Niche Profession', data=df, order=df['Niche Profession'].value_counts().index, palette='viridis')

# Set the title and axis labels
plt.title('Niche Profession Distribution', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Niche Profession', fontsize=12)
plt.show()  # Display the plot
```
