```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
# Replace 'college_fest_enquiry.csv' with the actual path or name of the uploaded file
df = pd.read_csv('/content/drive/MyDrive/Tailor AI/College Data Upated.csv')

# Preview the data to understand its structure
print(df)  # This will print the entire dataset
```

```python
# Count the occurrences of each niche profession
profession_count = df['Niche Profession'].value_counts()

# Create a lollipop chart for niche profession distribution
plt.figure(figsize=(10, 6))
plt.stem(profession_count.index, profession_count.values, linefmt='b-', markerfmt='bo', basefmt='b')
plt.title('Niche Profession Distribution in College Fest Enquiry', fontsize=16)
plt.xlabel('Niche Profession', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Import necessary libraries
from wordcloud import WordCloud

# Count the occurrences of each niche profession
profession_count = df['Niche Profession'].value_counts()

# Create a dictionary for the word cloud
word_freq = dict(zip(profession_count.index, profession_count.values))

# Generate a word cloud
plt.figure(figsize=(10, 10))
wordcloud = WordCloud(width=800, height=400, background_color='black').generate_from_frequencies(word_freq)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title('Niche Profession Word Cloud', fontsize=16)
plt.show()
```
