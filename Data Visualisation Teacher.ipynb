```python
import pandas as pd

# Load the CSV file
# Replace 'teachers.csv' with the actual path or name of the uploaded file
df = pd.read_csv('/content/drive/MyDrive/Tailor AI/Teacher Data Updated.csv')

# Preview the data to understand its structure
print(df)  # This will print the first 5 rows of the dataset
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Count the occurrences of each niche profession
profession_count = df['Niche Profession'].value_counts()

# Create a pie chart for niche profession distribution
plt.figure(figsize=(8, 8))
plt.pie(profession_count, labels=profession_count.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Niche Profession Distribution of Teachers', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is drawn as a circle
plt.show()
```
